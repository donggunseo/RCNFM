```python
subj_type_dict = {}
obj_type_dict = {}
class_dict = {}
ner_dict = []
for item in data1:
    if item['subj_type'] not in subj_type_dict:
        subj_type_dict[item['subj_type']] = 1
    else:
        subj_type_dict[item['subj_type']] += 1
    if item['obj_type'] not in obj_type_dict:
        obj_type_dict[item['obj_type']] = 1
    else:
        obj_type_dict[item['obj_type']] += 1
    if item['relation'] not in class_dict:
        class_dict[item['relation']] = 1
    else:
        class_dict[item['relation']] += 1
    for x in item['stanford_ner']:
        if x not in ner_dict:
            ner_dict.append(x)
    
```

```python
relation_type_dict = {k:{} for k in class_dict.keys()}
for item in data:
    txt = (item['subj_type'], item['obj_type'])
    if item['relation']=='no_relation':
        continue
    if txt not in relation_type_dict[item['relation']]:
        relation_type_dict[item['relation']][txt]=1
    else:
        relation_type_dict[item['relation']][txt]+=1
```

```python
for k,v in relation_type_dict.items():
        print(k)
        print(v)
        print('__________')
```

```python
entity_pair = []
for subj in subj_type_dict.keys():
    for obj in obj_type_dict.keys():
        entity_pair.append((subj,obj))
```

```python
entity_pair_relation = {k:[] for k in entity_pair}
for k,v in relation_type_dict.items():
    for e in v.keys():
        entity_pair_relation[e].append(k)
```
